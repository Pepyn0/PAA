````python
class Graph:

    def __init__(self, vertices):
        self.V = vertices
        self.graph = []


    def addAresta(self, ponto1, ponto2, peso):
        self.graph.append([ponto1, ponto2, peso])



    def find(self, representante, i):
        if representante[i] == i:
            return i
        return self.find(representante, representante[i])

    def union(self, representante, rank, x, y):
        xroot = self.find(representante, x)
        yroot = self.find(representante, y)

        if rank[xroot] < rank[yroot]:
            representante[xroot] = yroot
        elif rank[xroot] > rank[yroot]:
            representante[yroot] = xroot

        else:
            representante[yroot] = xroot
            rank[xroot] += 1

    def KruskalMST(self):

        self.graph = sorted(self.graph, key=lambda item: item[2])

        indiceAresta = 0
        indiceResult = 0
        result = []
        representante = []
        rank = []

        for node in range(self.V):
            representante.append(node)
            rank.append(0)

        while indiceResult < self.V - 1:

            ponto1, ponto2, peso = self.graph[indiceAresta]
            indiceAresta = indiceAresta + 1
            x = self.find(representante, ponto1)
            y = self.find(representante, ponto2)

            if x != y:
                indiceResult = indiceResult + 1
                result.append([ponto1, ponto2, peso])
                self.union(representante, rank, x, y)
            # Else discard the edge

        minimumCost = 0
        print("Pontes a serem construidas:")
        for ponto1, ponto2, weight in result:
            minimumCost += weight
            print("%d -- %d == %d" % (ponto1, ponto2, weight))
        print("MST ==" , minimumCost)
````
